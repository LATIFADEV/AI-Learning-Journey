```python
import pandas as pd
import numpy as np

# Création du dataset
data = {
    "id_transaction": range(1, 21),
    "date_vente": pd.date_range(start="2024-01-05", periods=20, freq="15D"),
    "ville": [
        "Marrakech", "Casablanca", "Rabat", "Marrakech", "Tanger",
        "Casablanca", "Rabat", "Marrakech", "Agadir", "Fes",
        "Marrakech", "Casablanca", "Rabat", "Tanger", "Agadir",
        "Fes", "Marrakech", "Casablanca", "Rabat", "Agadir"
    ],
    "type_bien": [
        "Appartement", "Villa", "Appartement", "Maison", "Appartement",
        "Villa", "Maison", "Appartement", "Villa", "Maison",
        "Appartement", "Villa", "Maison", "Appartement", "Villa",
        "Maison", "Appartement", "Villa", "Maison", "Appartement"
    ],
    "surface_m2": [
        80, 200, 75, 120, 65, 220, 110, 85, 250, 130,
        90, 210, 115, 70, 240, 140, 95, 230, 105, 88
    ],
    "prix_vente": [
        800000, 3200000, 750000, 1200000, 680000,
        3500000, 1100000, 820000, 4000000, 1300000,
        900000, 3300000, 1150000, 700000, 4200000,
        1350000, 950000, 3400000, 1080000, 870000
    ],
    "nombre_pieces": [
        3, 6, 3, 4, 2, 7, 5, 3, 8, 5,
        3, 6, 5, 2, 7, 5, 4, 6, 5, 3
    ]
}

df = pd.DataFrame(data)
```

```python
# Création de la colonne prix_m2
df["prix_m2"] = df["prix_vente"] / df["surface_m2"]
```

```python
# Calcul de la moyenne (en ignorant les NaN)
moyenne_prix_m2 = df["prix_m2"].mean()

# Remplacement des valeurs nulles
df["prix_m2"] = df["prix_m2"].fillna(moyenne_prix_m2)

print(df["prix_m2"])
```

```python
import numpy as np

df.loc[2, "prix_m2"] = np.nan
df.loc[5, "prix_m2"] = np.nan
```

```python
df["prix_m2"] = df["prix_m2"].fillna(df["prix_m2"].mean())
```

```python
# Supprimer les lignes sans numéro de transaction
df = df.dropna(subset=["id_transaction"])
```

```python
print(df.isnull().sum())
print(df.shape)
```

```python
# Moyenne, médiane et écart-type
moyenne = df["prix_vente"].mean()
mediane = df["prix_vente"].median()
ecart_type = df["prix_vente"].std()

print("Moyenne :", moyenne)
print("Médiane :", mediane)
print("Écart-type :", ecart_type)
```

```python
# Prix minimum et maximum
prix_min = df["prix_vente"].min()
prix_max = df["prix_vente"].max()

print("Prix minimum :", prix_min)
print("Prix maximum :", prix_max)
```

```python
# Coefficient de variation
coef_variation = ecart_type / moyenne

print("Coefficient de variation :", coef_variation)
```

```python
# Prix moyen par type de bien
prix_moyen_type = df.groupby("type_bien")["prix_vente"].mean()

print(prix_moyen_type)
```
