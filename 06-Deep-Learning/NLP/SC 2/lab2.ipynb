```python
# Dictionnaire de correspondance des étiquettes
label_names = {0: "World", 1: "Sports", 2: "Business", 3: "Sci/Tech"}

# Affichage d'un exemple du jeu d'entraînement
sample = dataset["train"][0]

print(f"Texte brut : \n{sample['text']}\n")
print(f"ID du Label : {sample['label']}")
print(f"Nom du Label : {label_names[sample['label']]}")
```

```python
# Mélange des données et sélection d'un sous-ensemble
train_sub = dataset["train"].shuffle(seed=42).select(range(5000))
test_sub = dataset["test"].shuffle(seed=42).select(range(1000))

print(f"Taille de notre ensemble d'entraînement réduit : {len(train_sub)} exemples")
print(f"Taille de notre ensemble de test réduit : {len(test_sub)} exemples")
```
