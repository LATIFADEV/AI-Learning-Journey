```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({
    "Nom": ["Amine", "Lina", "Youssef", "Salma", "Nora"],
    "Âge": [28, np.nan, 35, 30, np.nan],
    "Salaire": [5000, 6000, np.nan, 5500, 5200]
})
```

```python
df = pd.DataFrame({
    "Transaction": ["T1", "T2", "T3", "T4"],
    "Montant": [500, 1500, 2500, 800]
})
```

```python
df_filtered = df[df["Montant"] > 1000]
df_filtered
```

```python
df = pd.DataFrame({
    "Région": ["Nord", "Sud", "Nord", "Sud", "Est"],
    "Ventes": [1000, 1500, 2000, 1200, 1800]
})
```

```python
df_grouped = df.groupby("Région")["Ventes"].sum()
df_grouped
```

```python
clients = pd.DataFrame({
    "customer_id": [1, 2, 3],
    "Nom": ["Amine", "Lina", "Youssef"]
})
```

```python
commandes = pd.DataFrame({
    "customer_id": [1, 2, 1],
    "Commande": ["C1", "C2", "C3"]
})
```

```python
df_merged = pd.merge(clients, commandes, on="customer_id", how="inner")
df_merged
```

```python
df = pd.DataFrame({
    "Produit": ["A", "A", "B", "B"],
    "Région": ["Nord", "Sud", "Nord", "Sud"],
    "Ventes": [100, 150, 200, 250]
})
```

```python
pivot = df.pivot_table(values="Ventes", index="Produit", columns="Région", aggfunc="sum")
pivot
```
