```python
import numpy as np

# Tableau de températures (en °C)
temperatures = np.array([22, 24, 19, 23, 25, 21, 20])

# Calculs
moyenne = np.mean(temperatures)
mediane = np.median(temperatures)
ecart_type = np.std(temperatures)

# Affichage
print("Températures :", temperatures)
print("Moyenne :", moyenne)
print("Médiane :", mediane)
print("Écart-type :", ecart_type)
```

```python
# Tableau de données
data = np.array([10, 20, 30, 40, 50])

# Normalisation
moyenne = np.mean(data)
ecart_type = np.std(data)

data_normalise = (data - moyenne) / ecart_type

# Affichage
print("Données originales :", data)
print("Données normalisées :", data_normalise)
```

```python
# Création de deux tableaux
a = np.array([1, 2, 3, 4, 5])
b = np.array([1, 2, 0, 4, 9])

# Trouver les indices différents
indices_diff = np.where(a != b)

# Affichage
print("Indices différents :", indices_diff)
print("Valeurs dans a :", a[indices_diff])
print("Valeurs dans b :", b[indices_diff])
```

```python
# Création de matrices
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Multiplication matricielle
C = np.dot(A, B)

# Transposée
C_transpose = C.T

# Inverse (si possible)
C_inverse = np.linalg.inv(C)

# Affichage
print("Matrice C :\n", C)
print("Transposée de C :\n", C_transpose)
print("Inverse de C :\n", C_inverse)
```

```python
# Tableau
valeurs = np.array([5, 12, 8, 20, 3, 15])

# Condition : valeurs > 10
indices = np.where(valeurs > 10)

# Nouveau tableau filtré
valeurs_filtrees = valeurs[indices]

# Affichage
print("Valeurs originales :", valeurs)
print("Indices des valeurs > 10 :", indices)
print("Valeurs filtrées :", valeurs_filtrees)
```
