```python
from datetime import datetime
```

```python
class Produit:
    def __init__(self, nom, categorie, prix, quantite):
        self.nom = nom
        self.categorie = categorie
        self.prix = prix
        self.quantite = quantite

    def __str__(self):
        return f"{self.nom} | {self.categorie} | {self.prix} DH | Stock: {self.quantite}"
```

```python
class GestionnaireStock:
    def __init__(self):
        self.stock = {}
        self.historique_ventes = []
```

```python
    def ajouter_produit(self, nom, categorie, prix, quantite):
        if nom in self.stock:
            self.stock[nom].quantite += quantite
        else:
            self.stock[nom] = Produit(nom, categorie, prix, quantite)
```

```python
    def vendre_produit(self, nom, quantite):
        if nom not in self.stock:
            print("Produit introuvable")
            return

        produit = self.stock[nom]

        if produit.quantite < quantite:
            print("Stock insuffisant")
            return

        produit.quantite -= quantite
        montant = produit.prix * quantite
        date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

        self.historique_ventes.append((date, nom, quantite, montant))
```

```python
    def afficher_stock(self):
        for p in self.stock.values():
            print(p)
```

```python
    def calculer_ca(self):
        total = 0
        for vente in self.historique_ventes:
            total += vente[3]
        return total

    def rechercher_categorie(self, categorie):
        return [p for p in self.stock.values() if p.categorie == categorie]
```

```python
gs = GestionnaireStock()

gs.ajouter_produit("PC", "Informatique", 7000, 5)
gs.ajouter_produit("Souris", "Informatique", 100, 10)
gs.ajouter_produit("Chaise", "Mobilier", 500, 3)

gs.vendre_produit("PC", 1)

print("\n--- STOCK ---")
gs.afficher_stock()

print("\nCA:", gs.calculer_ca())

print("\nInformatique:")
for p in gs.rechercher_categorie("Informatique"):
    print(p)
```
